## Barrancabermeja

```python
datos = read_csv('Barrancabermeja_Lags.csv', encoding='latin-1', sep=";")
datos.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
knnImputer = KNNImputer (missing_values=np.nan, n_neighbors=2, 
                         weights="uniform", metric="nan_euclidean")
# Ajustamos el modelo e imputamos los missing values
knnImputer.fit(datos[['Hum_Rel_Max']])
datos['Hum_Rel_Max'] = knnImputer.transform(datos[['Hum_Rel_Max']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_1']])
datos['Hum_Rel_Max_1'] = knnImputer.transform(datos[['Hum_Rel_Max_1']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_2']])
datos['Hum_Rel_Max_2'] = knnImputer.transform(datos[['Hum_Rel_Max_2']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_3']])
datos['Hum_Rel_Max_3'] = knnImputer.transform(datos[['Hum_Rel_Max_3']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_4']])
datos['Hum_Rel_Max_4'] = knnImputer.transform(datos[['Hum_Rel_Max_4']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_5']])
datos['Hum_Rel_Max_5'] = knnImputer.transform(datos[['Hum_Rel_Max_5']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Max_6']])
datos['Hum_Rel_Max_6'] = knnImputer.transform(datos[['Hum_Rel_Max_6']]).ravel()

knnImputer.fit(datos[['Hum_Rel_Min']])
datos['Hum_Rel_Min'] = knnImputer.transform(datos[['Hum_Rel_Min']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_1']])
datos['Hum_Rel_Min_1'] = knnImputer.transform(datos[['Hum_Rel_Min_1']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_2']])
datos['Hum_Rel_Min_2'] = knnImputer.transform(datos[['Hum_Rel_Min_2']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_3']])
datos['Hum_Rel_Min_3'] = knnImputer.transform(datos[['Hum_Rel_Min_3']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_4']])
datos['Hum_Rel_Min_4'] = knnImputer.transform(datos[['Hum_Rel_Min_4']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_5']])
datos['Hum_Rel_Min_5'] = knnImputer.transform(datos[['Hum_Rel_Min_5']]).ravel()
knnImputer.fit(datos[['Hum_Rel_Min_6']])
datos['Hum_Rel_Min_6'] = knnImputer.transform(datos[['Hum_Rel_Min_6']]).ravel()
```
